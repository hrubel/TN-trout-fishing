```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
response = requests.get("http://www.eregulations.com/tennessee/fishing/region-1")
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
lakes = soup.find('div', attrs = {'id': "_idContainer048"})
```

```python
lakes
```

```python
names=lakes.find_all("h3")
```

```python
fish_list=lakes.find_all("ul")
```

```python
print(names)
```

```python
results=[]
for i, j in zip(names, fish_list):
    name = i.text
    for fish in j.find_all("li"):
        fish_name = fish.find("strong").text
        fish_desc = fish.text.split(": ")[1]
        results.append([name, fish_name, fish_desc])
```

```python
lake = pd.DataFrame(results, columns = ['Name', 'Species','Description']) 
```

```python
lake.head(50)
```

```python
lake["Region"]= "1"
```
